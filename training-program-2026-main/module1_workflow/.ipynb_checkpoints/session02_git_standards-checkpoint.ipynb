## 1. Conventional Commits Standard

Format: `<type>(<scope>): <description>`

| Type | Purpose |
|------|---------|
| `feat` | New feature |
| `fix` | Bug fix |
| `refactor` | Code restructure |
| `security` | Security patch |
| `test` | Add/update tests |
| `chore` | Config, deps |

````python
good = [
    'feat(auth): implement JWT refresh token rotation',
    'fix(payment): handle null card expiry gracefully',
    'security(api): sanitize user input on search endpoint',
    'refactor(user): extract validation to dedicated service',
]
bad = ['fix bug', 'update', 'wip', 'asdfgh', 'FINAL_v2']
print('✅ GOOD commits:')
for c in good: print(f'  {c}')
print('\n❌ BAD commits (will be rejected):')
for c in bad: print(f'  {c}')
````

## 2. Enterprise Gitflow

```
main ←── release ←── develop ←── feature/TICKET-ID-description
                              ←── hotfix/TICKET-ID-description
```
**Rules:**
- `main`: merge from release/hotfix only, always deployable
- `develop`: integration branch, CI must pass
- Feature branches: named `feature/PROJ-123-add-oauth`
- PRs require: 1 approval + all checks green + no unresolved comments
