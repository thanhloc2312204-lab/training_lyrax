## 1. Context Engineering Framework

AI produces bad code because of **missing context**. The CRATE framework:

```
C — Codebase context  (existing interfaces, types, patterns)
R — Role              ('You are a senior Python engineer at a fintech company')
A — Architecture      (which patterns, which constraints)
T — Task              (specific, measurable, unambiguous)
E — Examples          (show 1 good example of what you want)
```

**Prompt template:**
```
Role: [Senior engineer, specific domain]
Codebase context: [paste relevant interfaces/types]
Architecture constraints: [patterns to follow, libs to avoid]
Task: [precise requirement]
Expected output: [format: code only / with tests / with docs]
```

## 2. CodeX & Claude Workflow

| Tool | Best For | Shortcut |
|------|----------|----------|
| CodeX Composer | Multi-file tasks, scaffolding | Ctrl+I |
| CodeX Chat | Questions about current file | Ctrl+L |
| CodeX Inline | Quick local edits | Ctrl+K |
| Claude API | Complex reasoning, review | Programmatic |

**Best practices:**
- Commit a checkpoint before Composer runs
- Always review the diff before accepting
- Use `.codebasecontext` file to inject project rules automatically
- Break large tasks into 3-5 step chains
